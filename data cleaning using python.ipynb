```python
import pandas as pd
tech={
'Courses':["Spark","PySpark","Hadoop","Python","pandas"],
    'Fee' :[20000,25000,26000,22000,24000],
    'Duration':['30day','40days','35days','40days','60days'],
    'Discount':[1000,2300,1200,2500,2000]
  }
r=['r1','r2','r3','r4','r5']
df=pd.DataFrame(tech,index=r)
```

```python
df
```

```python
plt.bar(df['Courses'],df['Fee'],color='brown')
plt.title("Comparison")
plt.xlabel('course')
plt.ylabel('fees')
plt.show()
```

```python
plt.pie(df['Fee'],labels=df['Courses'],autopct="%1.1f%%")
```

```python
plt.plot(df['Courses'],df['Discount'])
plt.show()
```

```python
plt.plot(df['Courses'],df['Fee'])
```

```python

plt.figure(figsize=(10, 6)) 

plt.subplot(2, 2, 1)  
plt.bar(df['Courses'], df['Fee'], color='brown')
plt.title('Bar Chart')
plt.xlabel('Courses')
plt.ylabel('Fees')

plt.subplot(2, 2, 2)
plt.plot(df['Courses'], df['Fee'], color='blue', marker='o')
plt.title('Line Chart')
plt.xlabel('Courses')
plt.ylabel('Fees')


plt.subplot(2, 2, 3)
plt.scatter(df['Courses'], df['Fee'], color='green')
plt.title('Scatter Plot')
plt.xlabel('Courses')
plt.ylabel('Fees')

plt.subplot(2, 2, 4)
plt.barh(df['Courses'], df['Fee'], color='purple')
plt.title('Horizontal Bar Chart')
plt.xlabel('Fees')
plt.ylabel('Courses')


plt.tight_layout() 

plt.show()
```
